```python
# ex 1
# a
reset()
x = var('x')
y = function('y')(x)
eq = diff(y, x) == 2 * x * (1 + y*y)
eq
```

```python
show(eq)
```

```python
desolve(eq,y)
```

```python
show(desolve(eq, y))
```

```python
sol = desolve(eq, y)
_C = var('_C')
sol.substitute(_C == 1)
```

```python
[soll]=solve(sol,y(x))
f(x,_C)=soll.rhs()
f(x,_C)
```

```python
plot(f(x,1/2),x,-pi/2,pi/2,ymin=-10,ymax=10)
```

```python
#b
reset()
x=var('x')
y=function('y')(x)
eqd=(x*x-1)*diff(y,x)+2*x*y*y==0
f(x,_C)=desolve(eqd,y)
show(f(x,_C))
```

```python
yy=var('yy')
f(x,_C).substitute(y(x)==yy)
_C=var('_C')
f(x,yy,_C)=f(x,_C).substitute(y(x)==yy)
f(x,yy,_C)
```

```python
implicit_plot(f(x,yy,0),(x,-25,25),(yy,-25,25))
```

```python
#c
reset()
x=var('x')
y=function('y')(x)
eqd=(2*x*x*diff(y,x))-x*x-y*y==0
f(x,_C)=desolve(eqd,y)
show(f(x,_C))
```

```python
yy=var('yy')
f(x,_C).substitute(y(x)==yy)
_C=var('_C')
f(x,yy,_C)=f(x,_C).substitute(y(x)==yy)
f(x,yy,_C)
```

```python
implicit_plot(f(x,yy,1),(x,-25,25),(yy,-25,25))
```

```python
#2.a)
reset()
x=var('x')
y=function('y')(x)
eq=diff(y,x)==1+y^2
_C=var('_C')
[sol]=solve(desolve(eq,y,ics=[0,1]),y(x))
sol=sol.rhs()
g=plot(sol,x,-2,2,ymin=-10,ymax=10)
show(g)
```

```python
#2.d)
reset()
x=var('x')
y=function('y')(x)
edq=diff(y,x,2)-5*diff(y,x)+4*y==0
f(x)=desolve(edq,y,ics=[0,5,8])
show(f(x))
```

```python
plot(f(x),x,-15,15,ymin=-10,ymax=10)
```

```python
#3.a-b)
reset()
x,k=var('x,k')
y=function('y')(x)
edq=diff(y,x)+k/x*y==x^3
_C=var('_C')
f(x,k)=desolve(edq,y,ics=[1,0],ivar=x)
show(f(x,1))
g=plot(f(x,1),x,-2,2,ymin=-20,ymax=20)
show(g)
```

```python
#3.c)
reset()
x,k=var('x,k')
y=function('y')(x)
edq=diff(y,x)+k/x*y==x^3
_C=var('_C')
f(x,k,_C)=desolve(edq,y,ivar=x)
g=plot(f(x,1,-5),x,-2,2,ymin=-20,ymax=20)
for i in range (-4,5):
    gx=plot(f(x,1,i),x,-2,2,ymin=-20,ymax=20)
    g=g+gx
show(g)
```
